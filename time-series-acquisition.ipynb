```python
def get_people_data(url):
    
    # creating an empty list of people
    people_data = []
    
    # while url is not none this code will keep repeating
    while url:
        
        # make API requests
        response = requests.get(url)
        
        # creates a .json object from our response
        people = response.json()
        
        ## Add the current page's planet information to the `planet_data` list
        people_data.extend(people['results'])
        
        # # Get the next URL from the response
        url = people['next']
        
        #return dataframe
    return people_data
```

```python
people_data = get_people_data('https://swapi.dev/api/people/')
peeps = pd.DataFrame(people_data)
```

# Do the same thing, but for planets.

```python
import requests
import pandas as pd

def get_planets_data(url):
    
    # creating an empty list
    planet_data = []
    
    # while url is not none this code will keep repeating
    while url:
        
        
        response = requests.get(url)
        
        # creates a .json object from our response
        planets = response.json()
        
        ## Add the current page's planet information to the `planet_data` list
        planet_data.extend(planets['results'])
        
        # # Get the next URL from the response
        url = planets['next']
    return planet_data
```

```python
planets_data = get_planets_data("https://swapi.dev/api/planets/")
planets = pd.DataFrame(planets_data)
```

```python
planets.head()
```

# Extract the data for starships.

```python
# this is a copy paste function
def get_starship_data(url):
    ship_data = []
    while url:
        response = requests.get(url)
        ship = response.json()
        ship_data.extend(ship['results'])
        url = ship['next']
    return ship_data
```

```python
starship_data = get_starship_data('https://swapi.dev/api/starships/')
```

```python
ships = pd.DataFrame(starship_data)
```
